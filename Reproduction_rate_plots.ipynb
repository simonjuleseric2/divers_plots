```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def define_ytick(yt0):
    yt2=[]
    for yt in yt0:
        if np.sign(yt)>0:
            sign_yt='+'
        else:
            sign_yt='-'
        yt=int(round(abs(yt)))
        #yt.astype(int)
        #print(yt)
        #print(int(round(yt)))
        yt_new=sign_yt+str(yt)+' %'
        yt2.append(yt_new)
        #yt2=np.array(yt2)
        #yt2.astype(int)
    return yt2


# Define 2 different reproduction number:
r1=0.94 

r2=1.02

#start case at day0
start_case=1000

#follow up span:
num_days=90

#days vector:
x1=np.linspace(0, num_days, num_days+1, dtype=int)
print(x1)

#New cases per day:
y1=np.power(r1, x1)*start_case
y2=np.power(r2, x1)*start_case

#New cases per day (percent):
y12=(y1-start_case)/(start_case/100)
y22=(y2-start_case)/(start_case/100)

```

## Plot the Results:
<br>
<br>

```python
fig, ax = plt.subplots()

ax.bar(x1, y22, color='firebrick')
ax.bar(x1, y12, color='royalblue')


# Formatting x labels
#plt.xticks(rotation=90)
plt.tight_layout()
# Use absolute value for y-ticks
ticks =  ax.get_yticks()
tick2=define_ytick(ticks)
#ax.set_yticklabels([int(abs(tick)) for tick in ticks])

#ax.set_yticklabels(['{:,.0f}%'.format(x) for x in vals])
ax.set_yticklabels(tick2)
plt.legend(['r=1.02', 'r=0.94'])
plt.ylabel('change in new cases relative to baseline')
plt.xlabel('infection cycles (0=baseline)')

plt.savefig('rts_news3.png', bbox_inches='tight')

plt.show()
```
